```python
import pandas as pd

columns = [('wandb_export_2025-01-19T10_27_57.060-08_00.csv', ' - eval/3_pitch_time_acc', 'pitch_acc', 'high vs low pitch'),
           ('wandb_export_2025-01-19T10_28_05.322-08_00.csv', ' - eval/3_rms_electrode_acc', 'rms_electrode_acc', 'high vs low volume'),
           ('wandb_export_2025-01-19T10_28_23.231-08_00.csv', ' - eval/3_speech_time_acc', 'speech_acc', 'speech vs non-speech')]

# Read the CSV file
for figure_index in range(len(columns)):
    df = pd.read_csv(columns[figure_index][0])

    # Extract the columns
    x = df['Step']
    y = df['wc_bug5_seq_di64_t24_dm192_nh12_nl10_dr0.2_optMuon_eiZ_bs100_wd0.0_lr0.0015_lrm0.0015_r1'+columns[figure_index][1]]

    # Create the plot
    plt.figure(figsize=(2, 2))
    plt.rcParams.update({'font.family': 'Arial', 'font.size': 12})
    plt.plot(x, y, color='black', linewidth=2)
    plt.xlabel('train step')
    plt.ylabel('accuracy')
    plt.grid(True)
    plt.ylim(0.5, .95)
    plt.xscale('log')
    plt.xlim(0, 70000)
    plt.title(columns[figure_index][3], fontsize=12)
    plt.xticks([100, 1000, 10000], ['.1k', '1k', '10k'])
    plt.yticks([0.5, 0.6, 0.7, 0.8, 0.9])
    plt.savefig(columns[figure_index][2]+'.pdf', bbox_inches='tight')
```

```python
import pandas as pd

columns = [('wandb_export_2025-01-19T11_58_50.253-08_00.csv', ' - train_loss', 'train_loss', 'train loss'),
           ('wandb_export_2025-01-19T11_59_06.847-08_00.csv', ' - test_loss', 'test_loss', 'test loss')]
axis_limits = [(0, 5), (0, 5), (5, 14)]

plt.figure(figsize=(3, 2))
plt.rcParams.update({'font.family': 'Arial', 'font.size': 12})

# test loss
df = pd.read_csv(columns[1][0])
x = df['Step']
y = df['wc_bug5_seq_di64_t24_dm192_nh12_nl10_dr0.2_optMuon_eiZ_bs100_wd0.0_lr0.0015_lrm0.0015_r1'+columns[1][1]]
plt.plot(x, y, color='black', linewidth=3, label='test')

# train loss
df = pd.read_csv(columns[0][0])
x = df['Step']
y = df['wc_bug5_seq_di64_t24_dm192_nh12_nl10_dr0.2_optMuon_eiZ_bs100_wd0.0_lr0.0015_lrm0.0015_r1'+columns[0][1]]
plt.plot(x, y, color='gray', linewidth=3, label='train')

plt.ylabel('loss')
plt.xlabel('train step')
plt.grid(True)
plt.ylim(0, 5)
plt.xscale('log')
plt.xlim(0, 70000)
plt.xticks([100, 1000, 10000], ['.1k', '1k', '10k'])
plt.legend()
plt.savefig('train_test_loss.pdf', bbox_inches='tight')
```
